```python
#Writting a function to create pie chart for various variables, please see folllowing details of parameters
#data_set: Pass data set name which should be used for chart, example data_file
#var_name: variable name on which chart should be created, example "gender"
#color_list: provide colors list which should be used for chart, example ["blue", "pink"]
#chart_title: write chart title name, example "Gender based total Number of Billionaires"

def var_chart(data_set, var_name, color_list, chart_title):
    a = data_set[var_name].value_counts().head(7)
    index = a.index
    genders = a.values
    custom_colors = color_list
    plt.figure(figsize=(4, 4))
    plt.pie(genders, labels=index, colors=custom_colors)
    central_circle = plt.Circle((0, 0), 0.4, color='white')
    fig = plt.gcf()
    fig.gca().add_artist(central_circle)
    plt.rc('font', size=10)
    plt.title(chart_title, fontsize=18)
    plt.show()
```

```python
#Gender based total number of Billionaires
var_chart(data_file, "gender", ["blue", "pink"], "Gender based total Number of Billionaires")
```
